```python
import sklearn
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
df.columns
```

```python
X = df.drop(columns=['median_house_value', '<1H OCEAN', 'INLAND', 'ISLAND', 'NEAR BAY','NEAR OCEAN'])
y = df['median_house_value']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
scaler = MinMaxScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

```python
model_LinearRegression = LinearRegression()
```

```python
model_LinearRegression.fit(scaled_X_train,y_train)
```

```python
import pickle

with open('model_LinearRegression.pkl','wb') as file:
    pickle.dump(model_LinearRegression,file)
```

```python
with open('model_LinearRegression.pkl','rb') as file:
    model_LinearRegression = pickle.load(file)
```

```python
predictions = model_LinearRegression.predict(scaled_X_test)
```

```python
# Evaluation
import math

print(mean_squared_error(y_test,predictions))
print(mean_absolute_error(y_test,predictions))
print(math.sqrt(mean_squared_error(y_test,predictions)))
print(r2_score(y_test,predictions))
```

```python
df['median_house_value'].describe()
```

## Result

```python
#mean of median_house_value in dataset
median_house_value_mean  = df['median_house_value'].mean()
median_house_value_mean
```

```python
#Mean of predicted median_house_value
prediction_mean = predictions.mean()
prediction_mean
```

```python
#Error in prediction
print(median_house_value_mean - prediction_mean)
```

## POLYNOMIAL REGRESSION

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
```

```python
model_polynomial = make_pipeline(PolynomialFeatures(3),LinearRegression())
model_polynomial
```

```python
model_polynomial.fit(scaled_X_train,y_train)
```

```python
predictions = model_polynomial.predict(scaled_X_test)
```

```python
# Evaluation
import math

print(mean_squared_error(y_test,predictions))
print(mean_absolute_error(y_test,predictions))
print(math.sqrt(mean_squared_error(y_test,predictions)))
print(r2_score(y_test,predictions))
```

```python
predictions.mean()
```

```python
#Error in prediction
print(median_house_value_mean - prediction_mean)
```
